```python
import csv
from collections import defaultdict
import pandas as pd

# Define a dictionary to store scores and counts
course_scores = defaultdict(lambda: {'score_sum': 0, 'review_count': 0})

# Assign sentiment values
sentiment_map = {
    'Positive': 1,
    'Negative': -1,
    'Neutral': 0,
    'Mixed': 0
}

# Read the CSV data
csv_file = 'output_data/final_result.csv'

with open(csv_file, 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    # Iterate through the rows (ignoring headers if present)
    for row in reader:
        course_id = row[2]  
        sentiment = row[-1]
        
        # Assign a score based on the sentiment
        score = sentiment_map.get(sentiment, 0)
        
        # Update score and count for the course
        course_scores[course_id]['club_id']=row[1]
        course_scores[course_id]['score_sum'] += score
        course_scores[course_id]['review_count'] += 1



# Calculate average scores
df_scored_course_review = []
for course_id, values in course_scores.items():
    avg_score = values['score_sum'] / values['review_count']
    print(f"Course ID: {course_id}, Club ID: {values['club_id']}, Average Sentiment Score: {avg_score:.2f}")
    row = {
        "course_id": course_id,
        "club_id": values['club_id'],
        "score" : avg_score
    }
    df_scored_course_review.append(row)

# Convert the list of dictionaries into a DataFrame
df_scored_course_review_df = pd.DataFrame(df_scored_course_review)

# Display the DataFrame
#print(df_scored_course_review_df)
# Save data frame
df_scored_course_review_df.to_csv("output_data/scored_course_review.csv", encoding='utf-8', index=False, quoting=csv.QUOTE_ALL)
```
